```python
#simple plot
from matplotlib import pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.plot([2.2,3.2,4.2,5.2],[6,7,8,9])
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.scatter([1,2,3,4],[5,6,7,8],label='apple')
plt.scatter([2.2,3.2,4.2,5.2],[6,7,8,9],label="mango")
plt.legend()
plt.xlabel('price')
plt.ylabel('cost')
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.bar([1,2,5,8],[4,5,6,6])
plt.bar([4,3,6,7],[2,3,4,5])
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.bar([1,2,5,8],[4,5,6,6],width=1,label='mango')
plt.bar([1,2,5,8],[4,5,6,6],width=0.5,color='r',label='kite')
plt.xlabel('goog')
plt.ylabel('mac')
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.hist([1,2,3,4,5,11,12,11,21,22,20,23,40,41,44,44],[0,10,20,30,40,50])
plt.show()
```

```python
plt.pie([1,1,1,1])
#total =1+1+1+1=4 
#part = 1/4 =25%
plt.show()
```

```python
plt.pie([10,2,2,6])
plt.show()
```

```python


plt.pie([10,2,2,6],labels=['fruit','hen','egg','milk'],autopct='%1.1i%%')
plt.title('breakfast')
plt.legend()
plt.show()
```
