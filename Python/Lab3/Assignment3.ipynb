```python
import numpy as np
# Q3
import matplotlib.pyplot as plt

# Generate array of 200 values between -pi and pi
gen_arr = np.linspace(-np.pi, np.pi, 200)

# Calculate sin, cos, and tan values
sin_values = np.sin(gen_arr)
cos_values = np.cos(gen_arr)
tan_values = np.tan(gen_arr)

# Plot sin, cos, and tan curves
plt.plot(gen_arr, sin_values, label='sin')
plt.plot(gen_arr, cos_values, label='cos')
plt.plot(gen_arr, tan_values, label='tan')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Functions')
plt.legend()
plt.show()
```

```python
# Q4
import numpy as np 
from numpy import linalg as LA

arr_1 = np.array([[1, 2], [3, 4]])
arr_2 = np.array([[2, 4], [6, 9]])
```

```python
matrix_product = np.dot(arr_1, arr_2)
print("Matrix multiplication with Dot Product:") 
print(matrix_product)
```

```python
eigenvalues_1, eigenvectors_1 = LA.eig(arr_1)
eigenvalues_2, eigenvectors_2 = LA.eig(arr_2)
print("Eigenvalues and eigenvectors for arr_1:")
print("Eigenvalues:", eigenvalues_1)
print("Eigenvectors:", eigenvectors_1)
print("Eigenvalues and eigenvectors for arr_2:")
print("Eigenvalues:", eigenvalues_2)
print("Eigenvectors:", eigenvectors_2)

```

```python
determinant_1 = LA.det(arr_1)
determinant_2 = LA.det(arr_2)
print("Determinant of arr_1:", determinant_1)
print("Determinant of arr_2:", determinant_2)
```

```python
inverse_1 = LA.inv(arr_1)
inverse_2 = LA.inv(arr_2)
print("Inverse of arr_1:")
print(inverse_1)
print("Inverse of arr_2:")
print(inverse_2)
```
